Activity 01

1. Load the dataset

```r
data <- read.csv("heart.csv")
```

2. Find the number of records

```r
count <- 0
for (i in 1:nrow(data)) {
    count <- count + 1
}
print(paste("Number of records:", count))
```

3. Print the summary of RestingBP

```r
min_val <- min(data$RestingBP)
max_val <- max(data$RestingBP)
mean_val <- mean(data$RestingBP)

summary <- paste("Min:", min_val, "Max:", max_val, "Mean:", mean_val)
print(summary)
```

4. Find the number of patients with “higher than mean” RestingBP

```r
mean_val <- mean(data$RestingBP)
count <- 0
for (i in 1:nrow(data)) {
    if (data$RestingBP[i] > mean_val) {
        count <- count + 1
    }
}
print(paste("Number of patients with higher RestingBP than the mean: ", count))
```

5. Find unique chest pain types.

```r
unique_chest_pain_types <- unique(data$ChestPainType)
print(unique_chest_pain_types)
```

6. Print the ages of the first 10 Females whose resting BP is greater than 140.

```r
count <- 0
for (i in 1:nrow(data)) {
    if (data$Sex[i] == "F" && data$RestingBP[i] > 140 && count < 10) {
        print(data$Age[i])
        count <- count + 1
    }
}
```

7. Write a function which takes a row of data as input and outputs the attribute

```r
get_attribute <- function(row, attribute) {
    return(row[[attribute]])
}

cholesterol_value <- get_attribute(data[1, ], "Cholesterol")
print(paste("Cholesterol:", cholesterol_value))
```

8. information in a structured manner. If cholesterol level is less than 200 then print low else high.

```r
check_cholesterol_level <- function(cholesterol) {
    if (cholesterol < 200) {
        return("Low")
    } else {
        return("High")
    }
}

for (i in 1:nrow(data)) {
    cholesterol_value <- data$Cholesterol[i]
    result <- check_cholesterol_level(cholesterol_value)
    print(paste("Row", i, "- Cholesterol Level:", result))
    
}
```

Activity 2

1. Consider the following Table
2. 2. Represent the above table in a pie chart
● Using the Rainbow color palette
● Title of the pie chart - “City Population”

```r
labels <- c("Colombo", "Kandy", "Galle", "Anuradhapura", "Trincomalee", "Jaffna")
values <- c(580000, 450000, 330000, 380000, 280000, 310000)

pie(values, labels = labels, main = "City Population", col = rainbow(length(labels)))
```

3. Represent the above pie chart, where it shows the percentage except the city as Labels

```r
percentages <- round(values / sum(values) * 100)
labels_with_percent <- paste(percentages, "%", sep = " ")

pie(values, labels = labels_with_percent, main = "Pie Chart with Percentages", col =
rainbow(length(labels)))
```
